```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("/content/globalterrorismdb_0718dist.csv",encoding='latin1')
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(axis=1)
```

```python
data.isna().sum()
```

```python
data['approxdate'].fillna('Null',inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data['related'].dropna(how='all',inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
del data['related']
```

```python
data.head()
```

```python
plt.figure(figsize=(18,7))
order=data.groupby(['country_txt'])['attacktype1'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(x=order.index,y=order.values,data=data)
plt.xticks(rotation=90)
plt.ylabel('Attacktype1')
plt.show()
```

```python
plt.figure(figsize=(18,7))
sns.countplot(x='iyear',hue='success',data=data)
plt.xticks(rotation=90)
plt.show()
```

```python

fig,ax=plt.subplots(1,2,figsize=(18,10))
order=data.groupby(['attacktype1_txt'])['attacktype1'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[0],x=order.index,y=order.values,data=data)
ax[0].set_xticklabels(order.index,rotation=90)
plt.ylabel('Attacktype1')
order=data.groupby(['attacktype1_txt'])['attacktype2'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[1],x=order.index,y=order.values,data=data)
plt.xticks(rotation=90)
plt.ylabel('Attacktype2')

plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(18,10))
order=data.groupby(['targtype1_txt'])['targtype1'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[0],x=order.index,y=order.values,data=data)
ax[0].set_xticklabels(order.index,rotation=90)
plt.ylabel('targtype1')
order=data.groupby(['targtype1_txt'])['targtype2'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[1],x=order.index,y=order.values,data=data)
plt.xticks(rotation=90)
plt.ylabel('Targettype2')

plt.show()
```

```python
fig,ax=plt.subplots(1,3,figsize=(18,10))
order=data.groupby(['weaptype1_txt'])['weaptype1'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[0],x=order.index,y=order.values,data=data)
ax[0].set_xticklabels(order.index,rotation=90)
plt.ylabel('weapontype1')
order=data.groupby(['weaptype1_txt'])['weaptype2'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[1],x=order.index,y=order.values,data=data)
ax[1].set_xticklabels(order.index,rotation=90)
plt.ylabel('weapontype2')
order=data.groupby(['weaptype1_txt'])['weaptype3'].sum().sort_values(ascending=False).iloc[:20]
sns.barplot(ax=ax[2],x=order.index,y=order.values,data=data)
plt.xticks(rotation=90)
plt.ylabel('weapontype3')

plt.show()
```
